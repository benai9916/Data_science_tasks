```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hpc_df = pd.read_csv('final_one.csv')
```

```python
hpc_df.head()
```

```python
final_plot = hpc_df.groupby('year')

plt.figure(figsize=(18,10))
for c in final_plot['year'].unique():
    df = final_plot.get_group(c[0])
    sns.regplot(x=df['hpc_no'], y=df['GDP'])
    plt.title('Year 2010 to 2020', fontsize=14)
plt.show()
```

```python
list(sorted(hpc_df['year'].unique()))
```

```python
for country,year in zip(final_plot['year'].unique(), list(sorted(hpc_df['year'].unique()))):
    df = final_plot.get_group(country[0])
    col =['D', 'D', 'D', 'E', 'E', 'E', 'F', 'F', 'F', 'G', 'G', 'G',]
    plt.figure(figsize=(10,6))
    sns.regplot(x=df['hpc_no'], y=df['GDP'], color='g', truncate=False)
    plt.title('Year '+ str(year), fontsize=24)
    plt.xlabel('HCP', fontsize=14)
    plt.ylabel('$ GDP per capita', fontsize=14)
    plt.show()
```

```python
hpc_sum = pd.DataFrame(hpc_df.groupby(['country'])['hpc_no'].sum())
hpc_sum.head(8)
```

```python
ax = hpc_sum.sort_values('hpc_no', ascending=False).head(12).plot.bar(title='Top 12 countries with HPC')
ax.set_xlabel('Countries')
ax.set_ylabel('No of HPC')
plt.show()
```
